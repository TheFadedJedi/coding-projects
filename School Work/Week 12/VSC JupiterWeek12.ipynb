```python
import matplotlib.pyplot as plt
# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [15, 30, 25, 20]
# Creating a column (bar) graph
plt.bar(categories, values, color='skyblue', alpha=0.7, label='Sample Column Graph')
# Adding labels to the axes
plt.xlabel('Categories')
plt.ylabel('Values')
# Adding a title to the plot
plt.title('Sample Column Graph Example')
# Adding a legend to the plot
plt.legend()
# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
# Sample data
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 30, 20, 25]
# Colors for each category
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
# Exploding the 2nd category (Category B)
explode = (0, 0.1, 0, 0)
# Creating a pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
# Adding a title to the plot
plt.title('Sample Pie Chart Example')
# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [15, 30, 25, 20]
# Creating a bar graph
plt.bar(categories, values, color=['skyblue', 'lightgreen', 'lightcoral', 'orange'], alpha=0.7, label='Sample Bar Graph')
# Adding labels to the axes
plt.xlabel('Categories')
plt.ylabel('Values')
# Adding a title to the plot
plt.title('Sample Bar Graph Example')
# Adding a legend to the plot
plt.legend()
# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
# Reading data from the CSV file
data = pd.read_csv(r"C:\\Users\\aakmir\\Desktop\\data.csv")
# Extracting categories and values from the data
categories = data['Category']
values = data['Value']
# Define colors for each category
colors = ['blue','red','green','black']
# Creating a column (bar) graph
plt.bar(categories, values, color=colors, alpha=0.7, label='Sample Column Graph')
# Adding labels to the axes
plt.xlabel('Categories')
plt.ylabel('Values')
# Adding a title to the plot
plt.title('Sample Column Graph from Data File')
# Adding a legend to the plot
plt.legend()
# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Generating random data
np.random.seed(42)
num_points = 50
x_values = np.random.rand(num_points)
y_values = np.random.rand(num_points)
# Creating a scatter plot
plt.scatter(x_values, y_values, color='blue', marker='o', label='Random Data Points')
# Adding labels to the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# Adding a title to the plot
plt.title('Sample Scatter Plot Example')
# Adding a legend to the plot
plt.legend()
# Displaying the plot
plt.show()
```
