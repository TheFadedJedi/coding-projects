```python
# Example 1: Student Information
student_info = {
'name': 'John Doe',
'age': 22,
'major': 'Computer Science',
'GPA': 3.8
}
print(student_info['name'])
print(student_info['major'])
```

```python
# Example 2: Book Information
book = {
'title': 'To Kill a Mockingbird',
'author': 'Harper Lee',
'year': 1960,
'genre': 'Fiction',
'pages': 324
}
print(book['author'])
```

```python
# Example 3: Students Grades
student_grades = {
'John': 85,
'Sarah': 92,
'Michael': 78,
'Emily': 95,
'David': 88
}
print(student_grades['Sarah'])
```

```python
# Example 4: Creating a dictionary and accessing values
student = {
'name': 'John',
'age': 20,
'major': 'Computer Science'
}
print(student['name']) # Output: John
print(student['age']) # Output: 20
```

```python
# Example 5: Adding a new key-value pair to the dictionary
student['university'] = 'ABC University'
print(student) # Output: {'name': 'John', 'age': 20, 'major': 'Computer Science', 'university': 'ABC
University'}
```

```python
# Example 6: Modifying an existing value in the dictionary
student['age'] = 21
print(student) # Output: {'name': 'John', 'age': 21, 'major': 'Computer Science', 'university': 'ABC
University'}
```

```python
# Example 7: Iterating through the dictionary
for key, value in student.items():
print(f"{key}: {value}")
```

```python
# Example 1: Favorite Sports
favorite_sports = {'soccer', 'basketball', 'swimming', 'volleyball', 'tennis'}
print('volleyball' in favorite_sports)
```

```python
# Example 2: Common Interests
alice_interests = {'reading', 'hiking', 'painting', 'traveling'}
bob_interests = {'traveling', 'cooking', 'hiking', 'photography'}
common_interests = alice_interests.intersection(bob_interests)
print(common_interests)
```

```python
# Example 3: Unique Letters in a Word
word = 'mississippi'
unique_letters = set(word)
print(unique_letters)
```

```python
# Example 4: Performing Set Operations
set_a = {1, 2, 3, 4, 5}
set_b = {3, 4, 5, 6, 7}
union_set = set_a.union(set_b)
print(union_set)
intersection_set = set_a.intersection(set_b)
print(intersection_set)
difference_set = set_a.difference(set_b)
print(difference_set)
```

```python
# Example 5: Favorite Fruits
favorite_fruits = {'apple', 'banana', 'orange', 'mango'}
favorite_fruits.add('grapes')
print(favorite_fruits)
favorite_fruits.remove('banana')
print(favorite_fruits)
```

```python
# Example 6: Common Numbers
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}
common_numbers = set_a.intersection(set_b)
print(common_numbers)
```

```python
# Example 7: Union and Difference
set_x = {1, 2, 3, 4}
set_y = {3, 4, 5, 6}
union_set = set_x.union(set_y)
print(union_set)
difference_set = set_x.difference(set_y)
print(difference_set)
```

```python
# Example 8: Set Operations with Strings
set_a = set("programming")
set_b = set("python")
intersection_set = set_a.intersection(set_b)
print(intersection_set)
union_set = set_a.union(set_b)
print(union_set)
```
